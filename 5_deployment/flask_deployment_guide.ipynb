# Beginner-Friendly Flask Deployment Guide
This notebook will walk you through setting up a basic Flask application, running it locally, defining routes, rendering HTML templates, and handling form data.

## 1. Install Flask
To get started, install Flask using pip in your terminal:
```bash
pip install flask
```

## 2. Create a Basic Flask Application
Let's create a minimal Flask app instance.

````python
# app.py - Minimal Flask app
from flask import Flask
app = Flask(__name__)

@app.route('/')
def home():
    return "Hello, Flask!"
````

## 3. Run the Flask Development Server
To run your Flask app, use the following command in your terminal:
```bash
flask run
```
By default, the app will be available at http://127.0.0.1:5000/

## 4. Define Routes and Return Responses
You can define more routes to handle different URLs and return various responses.

````python
@app.route('/about')
def about():
    return "This is a simple Flask app for heart disease prediction."
````

## 5. Render HTML Templates
Flask can serve HTML files using the `render_template` function. First, create a folder named `templates` and add an HTML file (e.g., `index.html`).

````python
from flask import render_template

@app.route('/welcome')
def welcome():
    return render_template('index.html')
````

## 6. Handle Form Data
You can accept and process form data sent from the browser using Flask's request object.

````python
from flask import request

@app.route('/predict', methods=['POST'])
def predict():
    user_input = request.form.get('input_field')
    # Here you would process user_input and return a prediction
    return f"Received input: {user_input}"
````
