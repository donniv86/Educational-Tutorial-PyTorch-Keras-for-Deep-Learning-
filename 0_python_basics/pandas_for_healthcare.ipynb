# Data Analysis with Pandas (2026)

Learn to analyze, clean, and manipulate healthcare data using modern pandas features.

## 1. Creating and Exploring DataFrames
Work with tabular healthcare data.

```python
import pandas as pd
patient_data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [30, 45, 50, 40],
    'bp': [120, 135, 150, 135]
}
df = pd.DataFrame(patient_data)
print(df)
print('Mean age:', df['age'].mean())
print('Patients with BP > 130:')
print(df[df['bp'] > 130])
```

## 2. Data Cleaning and Transformation
Handle missing values, duplicates, and format data.

```python
df.loc[0, 'bp'] = None
print('With missing value:')
print(df)
df['bp'] = df['bp'].fillna(df['bp'].mean())
print('After filling missing values:')
print(df)
df = df.drop_duplicates()
print('After removing duplicates:')
print(df)
df['age'] = df['age'].astype(int)
print('Ages as integers:')
print(df)
```

## 3. GroupBy, Aggregation, and Pivot Tables
Summarize and reshape healthcare data.

```python
# Group by age group
df['age_group'] = pd.cut(df['age'], bins=[0, 40, 60], labels=['Young', 'Senior'])
grouped = df.groupby('age_group')['bp'].mean()
print('Mean BP by age group:')
print(grouped)
# Pivot table
pivot = df.pivot_table(values='bp', index='age_group', aggfunc='mean')
print('Pivot table:')
print(pivot)
```
