## 2. Comprehensions and Lambda Functions
Efficient data processing with list, dict, set comprehensions and lambdas.

```python
# List comprehension: Square numbers
squares = [x**2 for x in range(5)]
print('Squares:', squares)
# Dict comprehension: Age mapping
names = ['Alice', 'Bob', 'Charlie']
ages = [30, 45, 50]
age_dict = {name: age for name, age in zip(names, ages)}
print('Name to age mapping:', age_dict)
# Set comprehension: Even numbers
evens = {x for x in range(10) if x % 2 == 0}
print('Even numbers:', evens)
```
