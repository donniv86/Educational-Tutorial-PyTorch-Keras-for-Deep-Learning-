## 12. Biostatistics in Python: Concepts and Applications

Biostatistics is essential for analyzing and interpreting data. In this section, you'll learn how to perform key and advanced biostatistical analyses using Python, with scientific explanations and practical examples.

**Topics Covered:**
- Descriptive statistics (mean, median, mode, std, IQR)
- Probability distributions (normality, visualization)
- Hypothesis testing (t-test, chi-square, ANOVA)
- Correlation and regression
- Confidence intervals
- P-values and statistical significance
- Power analysis and sample size estimation
- Non-parametric tests (Mann-Whitney U, Wilcoxon, Kruskal-Wallis)
- Survival analysis (Kaplan-Meier, log-rank test)
- ROC curves and AUC for diagnostic tests
- Interpreting results in a scientific context

````python
# --- Descriptive Statistics, Distributions, and Hypothesis Testing ---
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# Example: Blood pressure data
bp_data = np.array([120, 135, 150, 110, 140, 130, 125, 145, 138, 132])
age = np.array([30, 45, 50, 40, 60, 55, 35, 65, 48, 52])

# ...existing code for descriptive stats, normality, t-test, correlation, regression, CI, chi-square, ANOVA...

# --- Power Analysis and Sample Size Estimation ---
from statsmodels.stats.power import TTestIndPower
# Calculate required sample size for detecting a difference in means
analysis = TTestIndPower()
# Example: effect size=0.8, alpha=0.05, power=0.8
sample_size = analysis.solve_power(effect_size=0.8, alpha=0.05, power=0.8)
print(f"Required sample size for 80% power (effect size=0.8): {sample_size:.1f}")

# --- Non-parametric Tests ---
# Mann-Whitney U test (2 independent groups)
group_a = np.array([120, 122, 119, 121])
group_b = np.array([130, 128, 135, 132])
mannu_stat, mannu_p = stats.mannwhitneyu(group_a, group_b)
print(f"Mann-Whitney U p-value: {mannu_p:.3f}")

# Wilcoxon signed-rank test (paired data)
pre = np.array([120, 130, 125, 140])
post = np.array([122, 128, 127, 138])
wilcoxon_stat, wilcoxon_p = stats.wilcoxon(pre, post)
print(f"Wilcoxon signed-rank p-value: {wilcoxon_p:.3f}")

# Kruskal-Wallis test (3+ groups)
group1 = np.array([120, 122, 119])
group2 = np.array([130, 128, 135])
group3 = np.array([140, 138, 142])
kruskal_stat, kruskal_p = stats.kruskal(group1, group2, group3)
print(f"Kruskal-Wallis p-value: {kruskal_p:.3f}")

# --- Survival Analysis (Kaplan-Meier, Log-rank Test) ---
from lifelines import KaplanMeierFitter, statistics
# Example survival data: durations and event observed (1=event, 0=censored)
durations = [5, 6, 6, 2, 4, 4, 3, 5, 7, 6]
event_observed = [1, 0, 1, 1, 1, 0, 1, 1, 0, 1]
kmf = KaplanMeierFitter()
kmf.fit(durations, event_observed=event_observed)
kmf.plot_survival_function()
plt.title('Kaplan-Meier Survival Curve')
plt.xlabel('Time (years)')
plt.ylabel('Survival Probability')
plt.show()

# Log-rank test between two groups
T1 = [5, 6, 6, 2, 4]
E1 = [1, 0, 1, 1, 1]
T2 = [4, 4, 3, 5, 7]
E2 = [0, 1, 1, 1, 0]
results = statistics.logrank_test(T1, T2, event_observed_A=E1, event_observed_B=E2)
print(f"Log-rank test p-value: {results.p_value:.3f}")

# --- ROC Curve and AUC for Diagnostic Tests ---
from sklearn.metrics import roc_curve, auc
# Example: true labels and predicted probabilities
y_true = np.array([0, 0, 1, 1, 0, 1, 0, 1, 1, 0])
y_scores = np.array([0.1, 0.4, 0.35, 0.8, 0.2, 0.7, 0.3, 0.9, 0.6, 0.2])
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()

# --- Scientific Explanation ---
# - Power analysis helps design studies with adequate sample size.
# - Non-parametric tests are used when data do not meet parametric assumptions.
# - Survival analysis estimates time-to-event outcomes (e.g., survival time).
# - ROC/AUC evaluate diagnostic test performance.
# - Always interpret results in the context of clinical/biological relevance.
````

## Epochs, Batch Size, and Their Effects in Gradient Descent

### Key Concepts
- **Epoch**: One complete pass through the entire training dataset.
- **Batch Size**: Number of samples processed before the model is updated.
- **Iteration**: One update of model weights (after each batch).

### Types of Gradient Descent
- **Batch Gradient Descent**: Uses the whole dataset as a single batch. Updates weights once per epoch. Slow but stable.
- **Stochastic Gradient Descent (SGD)**: Uses batch size of 1. Updates weights after each sample. Fast but noisy.
- **Mini-Batch Gradient Descent**: Uses batch size between 1 and dataset size (e.g., 32, 64, 128). Balances speed and stability.

### Effects on Learning
- **Epochs**: More epochs allow the model to learn more, but too many can cause overfitting.
- **Batch Size**: Small batch sizes make training noisier but can help escape local minima. Large batch sizes make training smoother but may require more memory.

### Illustrative Example
Suppose you have 200 samples, batch size 5, and 1000 epochs:
- Each epoch: 40 batches (200/5)
- Each batch: weights updated
- 1000 epochs: 40,000 weight updates

### Practical Tips
- Shuffle data between epochs for better generalization.
- Monitor loss and accuracy curves to decide when to stop training.
- Experiment with batch size and epochs to find optimal values for your dataset.

### References

- [Epoch vs Iterations vs Batch Size](https://towardsdatascience.com/epoch-vs-iterations-vs-batch-size-4dfb9c7ce9c9)

## Comparing Gradient Descent Approaches and Regularization

This section demonstrates how model performance changes with batch gradient descent (normal epochs) vs stochastic gradient descent, and how regularization helps prevent overfitting/underfitting. We'll use a simple linear regression example with synthetic data.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
np.random.seed(42)

# Generate synthetic data
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = 2 * X.squeeze() + 1 + np.random.normal(0, 2, 100)
# Overfitting example: add noise to training set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# Standardize features for SGD
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Batch Gradient Descent (LinearRegression)
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
mse_lr = mean_squared_error(y_test, y_pred_lr)

# Stochastic Gradient Descent (SGDRegressor)
sgd = SGDRegressor(max_iter=1000, tol=1e-3)
sgd.fit(X_train_scaled, y_train)
y_pred_sgd = sgd.predict(X_test_scaled)
mse_sgd = mean_squared_error(y_test, y_pred_sgd)

# Regularization: Ridge (L2)
sgd_reg = SGDRegressor(penalty='l2', alpha=0.1, max_iter=1000, tol=1e-3)
sgd_reg.fit(X_train_scaled, y_train)
y_pred_reg = sgd_reg.predict(X_test_scaled)
mse_reg = mean_squared_error(y_test, y_pred_reg)

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, label='Train Data', alpha=0.6)
plt.scatter(X_test, y_test, label='Test Data', alpha=0.6)
plt.plot(X_test, y_pred_lr, label=f'Batch GD (MSE={mse_lr:.2f})', color='blue')
plt.plot(X_test, y_pred_sgd, label=f'SGD (MSE={mse_sgd:.2f})', color='green')
plt.plot(X_test, y_pred_reg, label=f'SGD + L2 Reg (MSE={mse_reg:.2f})', color='red')
plt.title('Gradient Descent Approaches and Regularization')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

print('Batch Gradient Descent MSE:', mse_lr)
print('Stochastic Gradient Descent MSE:', mse_sgd)
print('SGD with L2 Regularization MSE:', mse_reg)
````

## Activation Functions and Optimizers in Deep Learning

This section explains common activation functions (ReLU, Sigmoid, Softmax) and optimizer techniques (SGD, Adam), with code and visualizations to help you understand their roles in neural networks.

````python
import numpy as np
import matplotlib.pyplot as plt

# Activation Functions
x = np.linspace(-10, 10, 200)

def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.plot(x, relu(x), label='ReLU')
plt.title('ReLU Activation')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x, sigmoid(x), label='Sigmoid', color='orange')
plt.title('Sigmoid Activation')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x, softmax(x), label='Softmax', color='green')
plt.title('Softmax Activation')
plt.xlabel('x')
plt.ylabel('Softmax(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Optimizer demonstration: Adam vs SGD
from sklearn.datasets import make_regression
from sklearn.neural_network import MLPRegressor

X, y = make_regression(n_samples=200, n_features=1, noise=20, random_state=42)

# SGD optimizer
sgd_model = MLPRegressor(hidden_layer_sizes=(10,), activation='relu', solver='sgd', max_iter=500, random_state=42)
sgd_model.fit(X, y)
sgd_pred = sgd_model.predict(X)

# Adam optimizer
adam_model = MLPRegressor(hidden_layer_sizes=(10,), activation='relu', solver='adam', max_iter=500, random_state=42)
adam_model.fit(X, y)
adam_pred = adam_model.predict(X)

plt.figure(figsize=(8, 5))
plt.scatter(X, y, label='Data', alpha=0.5)
plt.plot(X, sgd_pred, label='SGD Prediction', color='blue')
plt.plot(X, adam_pred, label='Adam Prediction', color='red')
plt.title('SGD vs Adam Optimizer Performance')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
````

## Other Optimizer Approaches in Deep Learning

Beyond SGD and Adam, several other optimizers are widely used:
- **RMSprop**: Adapts learning rates for each parameter by dividing by a moving average of recent gradients. Good for recurrent neural networks.
- **Adagrad**: Adapts learning rates based on the frequency of parameter updates. Works well for sparse data.
- **Adadelta**: An extension of Adagrad that seeks to reduce its aggressive, monotonically decreasing learning rate.
- **Nadam**: Combines Adam and Nesterov momentum for potentially faster convergence.

### When to Use Each Optimizer
- **SGD**: Simple, often used with momentum. Good for large datasets.
- **Adam**: Default for many tasks; works well in most cases.
- **RMSprop**: Good for RNNs and non-stationary objectives.
- **Adagrad/Adadelta**: Useful for sparse data or features.
- **Nadam**: Sometimes converges faster than Adam.

Below is a code example comparing Adam, RMSprop, Adagrad, and Nadam optimizers on a regression task.

````python
# Compare Adam, RMSprop, Adagrad, and Nadam optimizers using MLPRegressor
from sklearn.datasets import make_regression
from sklearn.neural_network import MLPRegressor
import numpy as np
import matplotlib.pyplot as plt

X, y = make_regression(n_samples=200, n_features=1, noise=20, random_state=42)

optimizers = ['adam', 'rmsprop', 'adagrad', 'nadam']
preds = {}
for opt in optimizers:
    model = MLPRegressor(hidden_layer_sizes=(10,), activation='relu', solver=opt, max_iter=500, random_state=42)
    model.fit(X, y)
    preds[opt] = model.predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Data', alpha=0.5)
colors = ['red', 'blue', 'green', 'purple']
for i, opt in enumerate(optimizers):
    plt.plot(X, preds[opt], label=f'{opt.capitalize()} Prediction', color=colors[i])
plt.title('Optimizer Comparison: Adam, RMSprop, Adagrad, Nadam')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
````

### Keras Example: Comparing Optimizers
The following code uses TensorFlow Keras to compare Adam, RMSprop, Adagrad, and Nadam optimizers on a simple regression task. Each optimizer is used to train a neural network, and the loss curves are plotted for comparison.
> Note: You need to have `tensorflow` installed. Run `pip install tensorflow` if needed.

````python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, RMSprop, Adagrad, Nadam

# Generate synthetic regression data
X = np.linspace(-1, 1, 200)
y = 3 * X + np.random.normal(0, 0.2, 200)

optimizers = {'Adam': Adam(), 'RMSprop': RMSprop(), 'Adagrad': Adagrad(), 'Nadam': Nadam()}
histories = {}

for name, opt in optimizers.items():
    model = Sequential([Dense(16, activation='relu', input_shape=(1,)), Dense(1)])
    model.compile(optimizer=opt, loss='mse')
    history = model.fit(X, y, epochs=50, batch_size=16, verbose=0)
    histories[name] = history.history['loss']

plt.figure(figsize=(10, 6))
for name, loss in histories.items():
    plt.plot(loss, label=name)
plt.title('Keras Optimizer Comparison: Loss Curves')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.grid(True)
plt.show()
````
