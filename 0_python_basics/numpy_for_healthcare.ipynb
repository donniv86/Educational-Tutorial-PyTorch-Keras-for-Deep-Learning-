# Scientific Computing with NumPy (2026)

Master NumPy for healthcare data: arrays, vectorized operations, random number generation, and modern features.

## 1. NumPy Arrays and Vectorized Operations
Efficiently store and process healthcare data.

```python
import numpy as np
bps = np.array([120, 135, 150, 110])
print('Blood pressures:', bps)
print('Mean BP:', np.mean(bps))
print('Max BP:', np.max(bps))
print('BP > 130:', bps > 130)
# Vectorized arithmetic
bps_mmHg = bps * 1.0  # conversion example
print('Converted BP:', bps_mmHg)
```

## 2. Random Number Generation and Reproducibility
Set seeds for reproducible experiments.

```python
np.random.seed(42)
random_bps = np.random.randint(100, 160, size=10)
print('Random blood pressures:', random_bps)
```

## 3. Modern NumPy Features (2026)
Broadcasting, advanced indexing, and new functions.

```python
# Broadcasting
ages = np.array([30, 45, 50, 40])
risk_scores = bps / ages
print('Risk scores:', risk_scores)
# Advanced indexing
high_bp_idx = np.where(bps > 130)
print('Indices of high BP:', high_bp_idx)
print('High BP values:', bps[high_bp_idx])
```
